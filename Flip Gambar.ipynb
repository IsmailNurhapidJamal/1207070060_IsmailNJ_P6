```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
# Membaca & menampilkan file gambar
img = imageio.imread("image/tank.jpg")
plt.imshow(img)
plt.axis('off')  # Menghilangkan sumbu x dan y
plt.show()
```

```python
#Mendapatkan/define resolusi dan tipe gambar
img_height = img.shape[0] #Pengambilan nilai tinggi citra 
img_width = img.shape[1] #Pengambilan nilai lebar citra
img_channel = img.shape[2] #Pengambilan jumlah channel citra 
img_type = img.dtype #Pengambilan tipe data (data type) dari citra
```

```python
#Membuat variable img_flip_horizontal  & img_flip_vertical untuk menampung
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)
```

```python
#Membuat perulangan yang akan mengiterasi semua nilai y dari 0
for y in range(0, img_height):
    #Perulangan yang akan mengiterasi semua nilai x dari 0
    for x in range(0, img_width):
        #Perulangan yang akan mengiterasi semua nilai c dari 0
        for c in range(0, img_channel):
            #Menghitung koordinat horizontal dari piksel di gambar yang telah diflip yang sesuai dengan piksel (y, x) pada gambar asli
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

```python
#Membuat perulangan yang akan mengiterasi semua nilai y dari 0
for y in range(0, img_height):
    #Perulangan yang akan mengiterasi semua nilai x dari 0
    for x in range(0, img_width):
        #Perulangan yang akan mengiterasi semua nilai c dari 0
        for c in range(0, img_channel):
            #Menghitung koordinat vertical dari piksel di gambar yang telah diflip yang sesuai dengan piksel (y, x) pada gambar asli
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

```python
#Menapilkan gambar dari img_flip_horizontal
plt.imshow(img_flip_horizontal)
plt.title("Flip Horizontal")
plt.show()
#Menapilkan gambar dari img_flip_vertical
plt.imshow(img_flip_vertical)
plt.title("Flip Vertical")
plt.show()
```
